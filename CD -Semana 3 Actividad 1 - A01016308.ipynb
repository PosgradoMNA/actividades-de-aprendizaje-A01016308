**Parte 1: Fundamentos de Bases de Datos**

Fundamentos de bases de datos y para ciencia de datos.  

Una base de datos es un programa donde se puede almacenar información de cualquier tipo de actividad en el mundo. Puede estar guardada de manera local (on-premise) o puede estar en la nube. El propósito de una base de datos es guardar cantidades masivas de información, con el fin de que podamos analizar de forma fácil y eficiente la data y crear predicciones que puedan ayudar a otras personas.

Para poder interactuar con las bases de datos, es importante que la información que se encuentre ahí esté normalizada. Sin coherencia, los datos no pueden predecir nueva información por lo que hay que asegurarnos que toda la información que se muestre ahí sea útil al limpiar la información, ya sea borrando la data que no nos sirve o en su caso reemplazando por algún valor que no afecte el desempeño de nuestro trabajo de análisis.

Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Los almacenes de datos son bases de datos muy grandes que contienen información de distintas fuentes. La ventaja es que el Data Warehouse las une todas para que podamos manipular la información de forma sencilla. Dado que es mucha información, la información debe pasar por una transformación para tener una base de datos estructurada. Antes se usaba mucho la metodología Extract, Transform, Load (ETL) dado que la cantidad de información no era tan grande, pero hoy día es diferente y las empresas pueden aprovechar servicios como el de Hadoop para hacer las transformación con el poder de procesamiento que estas empresas ofrecen, entonces actualmente se utiliza la metodología ELT, donde se carga primero la información en crudo, y la transformación sucede después.


**Parte 2: Selección y limpieza de los Datos en Python**

Funcionamiento del Código

1. En primer lugar, importamos las librerías a utilizar, guardamos el archivo .csv en la variable **input**, para después leerla y convertirla en **proyecto** a través de la variable *x*. Lo primero que hacemos es ver de cada una de las variables que tenemos: ID,X1...X23,Y, cuántos datos nulos tenemos. Copiaremos la base de datos a una base de datos **prueba** para trabajar sobre ella y no modificar los datos originales. En la impresión, podemos ver que todas las variables menos X1 presentan datos nulos, por lo que procederemos a analizar mejor esta información. Imprimiendo la variable **proyecto** podemos ver que el archivo en total maneja 30,000 columnas. 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

input="https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"

x=pd.read_csv(input)
proyecto=pd.DataFrame(x)
prueba=proyecto.copy()

print(proyecto,"\n")
print("Cantidad de renglones por variable con datos nulos \n")
print(prueba.isnull().sum())
```

2. Como paso 2, lo que haremos es usar la funcion dropna() con la variable **eliminar_renglones** con el objetivo de eliminar todos los renglones que contengan datos nulos. Como podemos ver en la impresión, 29,958 renglones quedan activos, lo que significa que los datos nulos se concentran en solamente 42 renglones. 



```python
eliminar_renglones=prueba.dropna()
print(eliminar_renglones)
```

3. Como tercer paso, haremos un archivo .csv que nos permita analizar más a fondo las 42 columnas que tienen datos nulos. Para esto, haremos uso de **eliminar_renglones** que previamente nos dió un DataFrame sin datos nulos. Nuevamente copiaremos la variable **prueba** para crear un nuevo DataFrame del archivo original para pruebas, a través de la variable **prueba_1**. Usaremos la función **pd.concat** para concatenar ambos archivos, lo que provocará que en la variable **concatenate** exista un DataFrame con los valores **SIN** datos nulos, repetidos. Para eliminarlos, usamos la funcion **.drop_duplicates** en la variable **datos_null**, lo que nos dará un archivo con solamente los 42 registros con datos nulos, y finalmente los exportamos a un .csv.

Con este archivo, podemos apreciar visualmente que la mayoría de los renglones tienen más de un dato faltante, y que dichos datos están agrupados en 5 conjuntos de ID relativamente cercanos, coloreados en la siguiente gráfica del lado izquierdo.



```python
prueba_1=prueba.copy()

concatenate=pd.concat([eliminar_renglones,prueba_1])
datos_null=concatenate.drop_duplicates(keep=False)

datos_null.to_csv("/Users/alexishema/Downloads/datos_null.csv")

print(datos_null)
```

4. Ahora que tenemos la información de los datos faltantes. Procederemos a hacer un ciclo **for** el cual nos va a ayudar a entender cómo están repartidos los outputs por variable. Por ejemplo: La variable **X2** tiene un output de 1 para hombre y 2 para mujer. Corriendo el ciclo, podemos ver que la respuesta 1 está en 18,112 renglones, mientras que la respuesta 2 está en 11,887, lo que nos da el total de 29,999 renglones que sabemos que si tienen información (solo un **X2** tenía datos nulos). Realizar este procedimiento en **proyecto** nos da una idea de la cantidad de outputs por variable que son incorrectos y que al mismo tiempo **no** son datos nulos. Por ejemplo, la variable **X4** nos dice el estado marital, siendo 1 casado, 2 soltero y 3 otros. Dentro del DataFrame hay 54 renglones con valor 0, lo cual no es un dato nulo, pero también es un factor que debemos de considerar al hacer la limpieza de datos, pues no es una respuesta válida.

```python

for x in range(0,23):
    c=proyecto.columns[x+2]
    A=str(c)
    print("Para la variable", c,"hay\n",proyecto[A].value_counts(),"\n")
  
print("Aquí termina el análisis por variable")
```

5. Aqui empezamos con la limpieza de datos por variable. 

Dado que para la variable **X2** solo hay un registro por llenar, y todos los registros con datos perdidos cercanos (color rosa) tienen el valor 1, se decide rellenar con 1.

Para la variable **X3** solo hay dos registros por llenar, para esta variable tanto la moda como la mediana tiene un valor de 2, por lo que se decide reemplazar con 2.

Para **X4** se tomó la misma decisión pues la moda y mediana tiene un valor de 2. Recordar que no podemos usar el promedio pues son variables categóricas, y una respuesta de ejemplo 2.3 no tendría sentido.

Para **X5** estamos hablando de la edad. El promedio de edad en la tabla de abajo (puros datos con faltantes) es de 36.1 La moda de **toda** la tabla (proyecto) es 29 y la mediana de 34. Dado que el valor de la mediana de todo el proyecto está más cercana al promedio de la tabla con datos faltantes, se decide sustituir con el valor de la mediana.

Las variables **X6-X11** representan históricos de pagos pasados, por lo que los vamos a tratar de la misma forma. Dado que esta columna representa el estado de estos mismos (si fueron liquidados o se deben), reemplazaremos los datos por moda, que en todos los casos es el valor 0.

Para las variables **X12-X17** estamos hablando de cantidades económicas, valores de factura. Estas cantidades están en números enteros y representan una moneda, por lo que utilizaremos el promedio para rellenar los datos faltantes, con el parámetro **skipna=True** para evitar que en el promedio se tomen en cuenta las celdas nulas.

Para las variables **X18-X23** estamos hablando de el valor de los pagos del mes previo. En el análisis anterior, podemos ver que el valor de estas celdas tiene una campana de Gauss inclinada a la izquierda, siendo los términos 0,1000,2000,3000,4000 y 5000 los más frecuentes, y la cola de los datos puede llegar hasta el valor de 85,000. Por esa razón es que se decidió utilizar la mediana para poder tener un número que parta la campana en 2 secciones de 50%.

Para la variable de salida **Y**, vemos que solo hay dos valores, el 0 y el 1. Tanto en la tabla **proyecto** como en la tabla con puros valores faltantes, la relación entre el 0 y 1 es 80:20, dado que faltan solamente 3 datos en el modelo, se decide reemplazar con moda, en este caso 0. Se anexa la misma tabla de valores faltantes para apoyo visual.



Se ejecuta el código para modificar los datos nulos y se vuelve a llamar la función **is.null().sum()** para corroborar que ya no hay más datos nulos en el DataFrame **data_clean**.

```python
data_clean=proyecto.copy()
data_clean["X2"].fillna(1,inplace=True)
data_clean["X3"].fillna(data_clean["X3"].mode()[0],inplace=True)
data_clean["X4"].fillna(data_clean["X4"].mode()[0],inplace=True)
data_clean["X5"].fillna(data_clean["X5"].median(),inplace=True)
data_clean["X6"].fillna(data_clean["X6"].mode()[0],inplace=True)
data_clean["X7"].fillna(data_clean["X7"].mode()[0],inplace=True)
data_clean["X8"].fillna(data_clean["X8"].mode()[0],inplace=True)
data_clean["X9"].fillna(data_clean["X9"].mode()[0],inplace=True)
data_clean["X10"].fillna(data_clean["X10"].mode()[0],inplace=True)
data_clean["X11"].fillna(data_clean["X11"].mode()[0],inplace=True)
data_clean["X12"].fillna(int(data_clean["X12"].mean(skipna=True)),inplace=True)
data_clean["X13"].fillna(int(data_clean["X13"].mean(skipna=True)),inplace=True)
data_clean["X14"].fillna(int(data_clean["X14"].mean(skipna=True)),inplace=True)
data_clean["X15"].fillna(int(data_clean["X15"].mean(skipna=True)),inplace=True)
data_clean["X16"].fillna(int(data_clean["X16"].mean(skipna=True)),inplace=True)
data_clean["X17"].fillna(int(data_clean["X17"].mean(skipna=True)),inplace=True)
data_clean["X18"].fillna(data_clean["X18"].median(),inplace=True)
data_clean["X19"].fillna(data_clean["X19"].median(),inplace=True)
data_clean["X20"].fillna(data_clean["X20"].median(),inplace=True)
data_clean["X21"].fillna(data_clean["X21"].median(),inplace=True)
data_clean["X22"].fillna(data_clean["X22"].median(),inplace=True)
data_clean["X23"].fillna(data_clean["X23"].median(),inplace=True)
data_clean["Y"].fillna(data_clean["Y"].mode()[0],inplace=True)

print(data_clean.isnull().sum())
```

6. Una vez que ya tenemos datos populados en todas las celdas, finalmente es hora de revisar si los valores que actualmente tiene cada variable tiene sentido, con base en la representación de cada columna.

**X3** va entre valores de 0,1,,3 y 4, siendo 4="otros". Vamos a cambiar todos los valores adicionales que aparecen (0,5,6) al valor 4 para calificar estas desviaciones como "otros".

**X4** va entre los valores 1,2 y 3, siendo 3="otros". Vamos a cambiar todos los valores adicionales que aparecen (0) al valor 3, para calificar esta desviación como "otros".

Todas las demás variables tienen outputs correctos.

```python
data_clean2=data_clean.copy()
data_clean2=data_clean2.replace({'X3':5},4)
data_clean2=data_clean2.replace({'X3':6},4)
data_clean2=data_clean2.replace({'X3':0},4)
data_clean2=data_clean2.replace({'X4':0},3)
```

7. Se vuelve a correr el código para ver los outputs de cada columna, verificando así que todos los outputs ya estén dentro de las salidas esperadas


```python
for x in range(0,23):
    c=data_clean2.columns[x+2]
    A=str(c)
    print("Para la variable", c,"hay\n",data_clean2[A].value_counts(),"\n")
  
print("Aquí termina el análisis por variable de la tabla limpia")
```

8. Finalmente, la tabla ya limpia se guarda en un archivo csv de manera local y en un nuevo DataFrame para su posterior uso.

```python
proyecto_limpio=data_clean2.copy()
proyecto_limpio.to_csv("/Users/alexishema/Downloads/Semana3Actividad1Limpio.csv")
print("Fin de la Parte 2")
```
