```python

import pandas as pd
import numpy as np
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
#Lectura del archivo csv de arriba, sin header 

cars = pd.read_csv(url, header=None)
cars.describe()
```

```python
#Incorporar los headers en nuestro DataFrame

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
cars.columns = headers
cars.head(10)
```

```python
cars.info()
```

```python
#reemplazar las columnas que digan ? con NaN

cars=cars.replace('?',np.NaN)
cars.head(10)
```

```python
#eliminar los renglones con price=NaN
cars=cars.dropna(subset=["price"], axis=0)
cars.head(20)
```

```python
#guardamos nuestro DataFrame a un archivo csv con una ruta local
cars.to_csv("/Users/alexishema/Downloads/automobile.csv", index=False)
```

```python
#vemos los tipos de formato para cada columna, más bien la info de cada columna
cars.dtypes
```

```python
#Se agregan los parámetros unique, top y frequence of the top
cars.describe(include = "all")
```
