**A continuación una descripción del contenido:**

This research employed a binary variable, default payment (Yes = 1, No = 0), as the response variable. This study reviewed the literature and used the following 23 variables as explanatory variables: 

X1: Amount of the given credit (NT dollar): it includes both the individual consumer credit and his/her family (supplementary) credit. 

X2: Gender (1 = male; 2 = female). 

X3: Education (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

X4: Marital status (1 = married; 2 = single; 3 = others). 

X5: Age (year).

X6 - X11: History of past payment. We tracked the past monthly payment records (from April to September, 2005) as follows: X6 = the repayment status in September, 2005; X7 = the repayment status in August, 2005; . . .;X11 = the repayment status in April, 2005. The measurement scale for the repayment status is: -1 = pay duly; 1 = payment delay for one month; 2 = payment delay for two months; . . .; 8 = payment delay for eight months; 9 = payment delay for nine months and above. 

X12-X17: Amount of bill statement (NT dollar). X12 = amount of bill statement in September, 2005; X13 = amount of bill statement in August, 2005; . . .; X17 = amount of bill statement in April, 2005. 

X18-X23: Amount of previous payment (NT dollar). X18 = amount paid in September, 2005; X19 = amount paid in August, 2005; . . .;X23 = amount paid in April, 2005.

##**3. Limpia los datos eliminando los registros nulos o rellena con la media de la columna**


De acuerdo con el análisis hecho, las variables se pueden clasificar en categóricas y numéricas, de la siguiente manera:

- Columnas con datos categóricos: X2, X3, X4, X6, X7, X8, X9, X10, X11, Y 
- Columnas con datos continuos: X1, X5, X12, X13, X14, X15, X16, X17, X18, X19, X20, X21, X22, X23


En esta parte del código vamos a **reemplazar por media los valores nulos**

```python
clean_creditos = creditos.copy()
moda = ['X1', 'X2', 'X3', 'X4', 'X5', 'X6', 'X7', 'X8', 'X9', 'X10',
       'X11', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20',
       'X21', 'X22', 'X23', 'Y'],
i=0
#Variables de Entrada CATEGORICAS
#Se usa la moda para llenar espacio vacíos, para ambos conjuntos de acuerdo con las instrucciones
for i in moda[i]:
    clean_creditos[i].fillna(value= clean_creditos[i].mean(), inplace = True)


print('Aquí se confirma que no haya datos faltantes')
clean_creditos.isnull().any() 
```

En esta parte del código vamos a **eliminar los valores nulos**, con este último ejercicio es que continuaremos el resto de la actividad.

```python
creditos2=creditos.copy()
creditos2.dropna(inplace=True)
creditos2.reset_index(inplace=True)
creditos2.shape
```

**Aquí se adjunta la descripción de las medidas de tendencia central y dispersión**

1. count: Es la cantidad de datos sin NA por columna.


2. mean: Es la suma de todos los datos, dividido entre la cantidad de datos, todos distintos a NA.


3. std: Desviación estándar, da una idea de qué tan dispersos o alejados se encuentran los datos del conjunto respecto del mean de cada columna.


4. min: Es el valor mínimo de cada columna.


5. 25%: Es el primer cuartil. Se ordena el conjunto de menor a mayor, este es el valor que se encuentra en el lugar (0.25 * count) de cada columna.


6. 50%: Este es el valor que se encuentra en el lugar (count * 0.50) de cada columna.


7. 75%: Este es el valor que se encuentra en el lugar (count * 0.75) de cada columna.


8. max: Es el valor máximo de cada columna.

##**6. Escala los datos, si consideras necesario**

```python
numericas = ['X1', 'X5', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23']

numericas_sin_escalar = creditos2[numericas]
std_scale = StandardScaler()
scaled_numeric = std_scale.fit_transform(numericas_sin_escalar)
#scaled_numeric2 = std_scale.transform(numericas_sin_escalar)
clean_creditos3 = creditos2.copy()
clean_creditos3[numericas] = pd.DataFrame(scaled_numeric)
clean_creditos3
```

##**7. Reduce las dimensiones con PCA, si consideras necesario.**
- Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%
- Indica la importancia de las variables en cada componente

```python
# Analiazamos la matriz de correlación del archivo antes de la escalación
corr_matrix = creditos2.corr()

sns.set(rc = {'figure.figsize':(17,10)})

sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='YlGnBu')
```

```python
components_amount = 5
pca = PCA(n_components=components_amount)
pca_data = pca.fit_transform(clean_creditos3[numericas])

print('Aquí indicamos la varianza de los datos explicada por cada componente seleccionado.\n')
print('     PC1       PC2        PC3          PC4       PC5')
print(pca.explained_variance_ratio_,'\n')
print('Aquí acumulamos la cantidad de varianza explicada por PCA\n')
print(sum(pca.explained_variance_ratio_),'\n')
print('Aquí indicamos la importancia de las variables en cada componente\n')
components_clean_creditos3 = pd.DataFrame(data=pca.components_.T, columns=[f'PC{i+1}' for i in range(components_amount)], index=numericas)
print(components_clean_creditos3)
```

##**8. Elabora los histogramas de los atributos para visualizar su distribución**

```python
def histogram_show(x):
  print(f'{x.name} Component Histogram')
  plt.hist(x)
  plt.show()

components_clean_creditos3.apply(histogram_show)
```
