```python
# Module 2 - Data Analysis with Python

# Importamos las librerías para trabajar en los ejemplos

import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
# URL del archivo a trabajar

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

# Vamos colocando el nombre de las columnas

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
coches = pd.read_csv(filename, names = headers)
coches.head()
```

```python
# Reemplazamos las celdas de "?" con NaN
coches.replace("?", np.nan, inplace = True)
coches.head(5)
```

```python
# Función .isnull para identificar los renglones con valores nulos
# True significa que sí hay valores nulos

sindata = coches.isnull()
sindata.head(5)
```

```python
#Contar la cantidad de valores nulos por columna, los que vengan con valor de True

for column in sindata.columns.values.tolist():
    print(column)
    print (sindata[column].value_counts())
    print("")  
```

```python
# Calcular el promedio para la columna de "normalized-losses"
promedio_nl = coches["normalized-losses"].astype("float").mean(axis=0)
print("Promedio de normalized-losses:", promedio_nl)
```

```python
# Para la misma columna, reemplazamos los valores en NaN con el promedio que acabamos de calcular

coches["normalized-losses"].replace(np.nan, promedio_nl, inplace=True)
```

```python
# Copiamos la misma metodología para la columna de bore

promedio_bore=coches['bore'].astype('float').mean(axis=0)
print("Promedio de bore:", promedio_bore)
coches["bore"].replace(np.nan, promedio_bore, inplace=True)
```

```python
# Para conocer los valores presentes en una columna, usamos la función .value_counts()

coches['num-of-doors'].value_counts()
```

```python
# Con .idxmax() conocemos el valor más frecuente en una columna
coches['num-of-doors'].value_counts().idxmax()
```

```python
# Como en esta columna no podemos reemplazar por promedio, pues es una variable categórica
# reemplazamos directo con el valor "four"

coches["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# Finalmente, eliminamos los renglones que aún tengan valores nulos

coches.dropna(subset=["price"], axis=0, inplace=True)

# Hacemos reset de los índices
coches.reset_index(drop=True, inplace=True)
coches.head(5)
```

```python
# Misma acción para horsepower

avg_horsepower = coches['horsepower'].astype('float').mean(axis=0)
print("Promedio horsepower:", avg_horsepower)
coches['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# Checamos el tipo de data para cada columna

coches.dtypes
```

```python
# Hacemos los cambios de tipo de variable dependiendo del análisis que hicimos arriba
# Ejemplo, bore y stroke son objeto, pero deben ser números reales por eso los cambiamos a float 

coches[["bore", "stroke"]] = coches[["bore", "stroke"]].astype("float")
coches[["normalized-losses"]] = coches[["normalized-losses"]].astype("int")
coches[["price"]] = coches[["price"]].astype("float")
coches[["peak-rpm"]] = coches[["peak-rpm"]].astype("float")
coches.dtypes
```

```python
# Estandarización de datos...colocar todas las variables dentro del mismo rango para que una, por defecto
# no tenga más peso que las otras, simplemente por su magnitud

coches['city-L/100km'] = 235/coches["city-mpg"]
coches.head()
```

```python
# Aplicar normalización mediante otras fórmulas que también son validas a otras columnas

coches['length'] = coches['length']/coches['length'].max()
coches['width'] = coches['width']/coches['width'].max()
```

```python
# Cambiamos la columna horsepower de objeto a entero
coches["horsepower"]=coches["horsepower"].astype(int, copy=True)
```

```python
# Graficamos los valores de la columna para ver cómo están distribuidos

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(coches["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Creamos los bins para poder armar las 3 categorías de clasificación

bins = np.linspace(min(coches["horsepower"]), max(coches["horsepower"]), 4)
bins
```

```python
# Los nombres de las nuevas clasificaciones 

group_names = ['Low', 'Medium', 'High']
```

```python
# Creamos la nueva columna horsepower-binned y otorgamos clasificación de acuerdo con el valor propio

coches['horsepower-binned'] = pd.cut(coches['horsepower'], bins, labels=group_names, include_lowest=True )
coches[['horsepower','horsepower-binned']].head(20)
```

```python
# Comprobamos como están organizados los datos y graficamos de acuerdo con la nueva clasificación

coches["horsepower-binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, coches["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(coches["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Se crea una nueva DataFrame llamado dummy_variable_1 para aplicar el one hot encoder en las variables categóricas

dummy_variable_1 = pd.get_dummies(coches["fuel-type"])
dummy_variable_1.head()
```

```python
# Se cambia el nombre de las variables para mejor reconocimiento

dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# Metemos las nuevas columnas al DataFrame original "coches" y eliminamos la columna original de tipo de fuel

coches = pd.concat([coches, dummy_variable_1], axis=1)

coches.drop("fuel-type", axis = 1, inplace=True)
coches.head()
```
